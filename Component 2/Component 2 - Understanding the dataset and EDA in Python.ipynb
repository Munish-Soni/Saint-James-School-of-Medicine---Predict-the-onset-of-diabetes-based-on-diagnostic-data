```python
'''

After Inspection, we can see that there are relatively small number of missing values (zeros) in 'BMI' & 'BloodPressure' column.

Therefore, dropping the data with null values should be the best approach.

'''
```

```python
# Creating a copy of the data, so the original data remains intact

diabetes_data = diabetes.copy()
```

```python
# Removing missing values from BMI column

diabetes_data['BMI'].replace(0, None, inplace=True)
diabetes_data.dropna(axis=0, inplace=True)
diabetes_data['BMI'] = diabetes_data['BMI'].astype('int64')
diabetes_data.info()
```

```python
# Removing missing values from BloodPressure column

diabetes_data['BloodPressure'].replace(0, None, inplace=True)
diabetes_data.dropna(axis=0, inplace=True)
diabetes_data['BloodPressure'] = diabetes_data['BloodPressure'].astype('int64')
diabetes_data.info()
```

```python
# As we know, Data imputation is performed on Null values & not zeroes, so we replace zeros with Null values in SkinThickness and Insulin columns 

diabetes_data['SkinThickness'].replace(0, None, inplace=True)
diabetes_data['Insulin'].replace(0, None, inplace=True)
```

```python
# Using IterativeImputer for imputation

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()

imputer.fit (diabetes)

diabetes_data = imputer.transform (diabetes_data)
```

```python
# Converting the output into a dataframe

from pandas import DataFrame
diabetes_data = DataFrame(diabetes_data, columns=diabetes.columns)
diabetes_data
```

```python
# Separating the predictor and Target Variables

X = diabetes_data.drop(columns=['Outcome'])
Y = diabetes_data['Outcome']
```

```python
# We will be using IsolationForest Algorithm for Outlier detection on the entire dataset

from sklearn.ensemble import IsolationForest

If = IsolationForest()

diabetes_data['anomaly_score'] = If.fit_predict(X)

diabetes_data
```

```python
# Total number of anomalies in the data

diabetes_data['anomaly_score'].value_counts()
```

```python
# Percentage of anomalies in the data

70/659
```

```python
# Now, since the anomalies are less than 20%, we will proceed with the Trimming approach on the data

diabetes_clean = diabetes_data[diabetes_data['anomaly_score'] == 1]

diabetes_clean.drop(columns = ['anomaly_score'], inplace=True)

diabetes_clean
```

```python
# Writing the clean data to an excel file

diabetes_clean.to_excel('diabetes_clean' + '.xlsx')
```
